```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph # framework that helps you design and manage the flow of tasks in your application using a graph structure
from IPython.display import Image, display
```

```python
# Agent state to keep the state while processing the nodes

class AgentState(TypedDict):  
    message : str 

def Gretter(state: AgentState) -> AgentState:
    """Simple node that adds a greeting message to the state"""
    state['message'] = "Hey " + state["message"] + ", how is your day going?"
    return state 
```

```python
graph = StateGraph(AgentState)
graph.add_node("Greetting", Gretter)
graph.set_entry_point("Greetting")
graph.set_finish_point("Greetting")
app = graph.compile()
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"message": "Bob"})
result["message"]
```
